```python
# 导入必要的函数
from DDAGI import automated_workflow, DrugDiscoveryAPI
import py3Dmol
from IPython.display import Image, HTML, display
import os

# 设置API密钥
os.environ["OPENAI_API_KEY"] = "sk-kiuwnsdtlpclsjguvgajhuqdgowypqhmgozbxhhnenucutdp"

# 指定疾病名称
disease = input("输入需要探索的疾病名称")

# 运行工作流
result = automated_workflow(disease)

# 显示分子图像
if result.get("molecule_image"):
    display(Image(filename=result["molecule_image"]))

# 显示优化的SMILES
print(f"优化后的分子SMILES: {result['optimized_smiles']}")

# 如果在Jupyter中，可以创建交互式的3D分子可视化
# (这部分在Jupyter中会更好地工作)
if result.get("structure_path") and result.get("optimized_smiles"):
    from rdkit import Chem
    from rdkit.Chem import AllChem
    import py3Dmol
    
    # 准备蛋白质和配体
    protein_path = result["structure_path"]
    smiles = result["optimized_smiles"]
    pocket_center = result["pocket_center"]
    
    # 创建配体3D结构
    mol = Chem.MolFromSmiles(smiles)
    mol = Chem.AddHs(mol)
    AllChem.EmbedMolecule(mol, AllChem.ETKDG())
    AllChem.UFFOptimizeMolecule(mol)
    
    # 保存为临时PDB文件
    from pathlib import Path
    import tempfile
    temp_dir = Path(tempfile.gettempdir())
    ligand_pdb = temp_dir / "temp_ligand.pdb"
    Chem.MolToPDBFile(mol, str(ligand_pdb))
    
    # 创建可视化
    view = py3Dmol.view(width=800, height=600)
    
    # 加载蛋白质
    with open(protein_path, 'r') as f:
        protein_data = f.read()
    view.addModel(protein_data, 'pdb')
    
    # 加载配体
    with open(ligand_pdb, 'r') as f:
        ligand_data = f.read()
    view.addModel(ligand_data, 'pdb')
    
    # 设置样式
    view.setStyle({'model': 0}, {'cartoon': {'color': 'spectrum'}})
    view.setStyle({'model': 1}, {'stick': {'colorscheme': 'cyanCarbon', 'radius': 0.2}})
    
    # 突出显示口袋
    x, y, z = pocket_center
    view.addSphere({'center': {'x': x, 'y': y, 'z': z}, 'radius': 3.0, 'color': 'yellow', 'opacity': 0.5})
    
    # 设置视图
    view.zoomTo()
    view
```
